```python
import numpy as np
import matplotlib.pyplot as plt
```

1.Para um sinal x(t) = cos(2πft) com duração de 5 s e frequência de amostragem fs = 44.1 kHz, faça o que se
pede nos itens abaixo:

(a) Gere gráficos de x(t) no domínio do tempo para f ∈ {500,5000,10000} Hz.

(b) Escreva um código para tocar os sinais gerados.

(c) Comente os resultados obtidos.

```python
duracao=0.01
fs=44.1e3
ts=duracao/fs
f1=500
f2=5000
f3=10000
time=np.arange(0,duracao, ts)
```

```python
y1=np.cos(2*np.pi * f1*time)
plt.plot(time, y1)

plt.xlabel("Tempo")
plt.ylabel("Amplitude")
```

```python
plt.plot(time, np.cos(2*np.pi * f2*time))

plt.xlabel("Tempo")
plt.ylabel("Amplitude")
```

```python
plt.plot(time, np.cos(2*np.pi * f3*time))

plt.xlabel("Tempo")
plt.ylabel("Amplitude")
```
