```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

## 1.Para um sinal x(t) = cos(2πft) com duração de 5 s e frequência de amostragem fs = 44.1 kHz, faça o que se pede nos itens abaixo:

(c) Comente os resultados obtidos.

### (a) Gere gráficos de x(t) no domínio do tempo para f ∈ {500,5000,10000} Hz.

```python
duracao=5
fs=44.1e3
ts=duracao/fs
f1=500
f2=4000
f3=10000
time=np.arange(0,duracao, ts)
```

```python
y1=np.cos(2*np.pi * f1*time)
plt.plot(time, y1)

plt.xlabel("Tempo")
plt.ylabel("Amplitude")
```

```python
y2 = np.cos(2*np.pi * f2*time)
plt.plot(time, y2)

plt.xlabel("Tempo")
plt.ylabel("Amplitude")
```

```python
y3 = np.cos(2*np.pi * f3*time)

plt.plot(time, y3)

plt.xlabel("Tempo")
plt.ylabel("Amplitude")
```

### (b) Escreva um código para tocar os sinais gerados.

```python
#Onda de 500Hz

Audio(y1, rate=fs)
```

```python
# Onda de 5000Hz

Audio(y2, rate=fs)
```

```python
# Onda de 1000Hz

Audio(y3, rate=fs)
```
