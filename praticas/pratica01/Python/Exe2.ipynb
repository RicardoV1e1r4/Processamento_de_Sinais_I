## 2.Para um chirp com frequência inicial f0 = 500 Hz, frequência final f1 = 10000 Hz, duração de 5 s e frequência de amostragem fs = 44.1 kHz, faça o que se pede nos itens abaixo:

(c) Comente os resultados obtidos.

```python
import numpy as np
import matplotlib.pyplot as plt
#from scipy.io import wavfile
from scipy.signal import chirp
from IPython.display import Audio
```

```python
# Parâmetros
pi = np.pi

duration=5
fs=44.1e3

f0=500
f1=10e3

time = np.arange(0, duration, duration/fs)
# time = np.linspace(0, duration, int(fs*duration), endpoint=False)
```

### (a) Gere gráficos de chirps no domínio do tempo, com varreduras de frequência lineares, quadráticas e logarítmicas.

```python
linear_chirp=chirp(time, f0=f0, f1=f1, t1=duration, method="linear")
```

```python
alpha = (f1 - f0)/(duration*2)

lin_chirp = np.sin(2*pi*(alpha*time + f0)*time)
```

```python
figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.plot(time, linear_chirp)
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Amplitude")
ax1.grid(True)

ax2.plot(time, linear_chirp)
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("Amplitude")
ax2.grid(True)

plt.show()
```

```python
# Quadratic chirp
# Sabendo que c = f0 e considerando a = 1

a = (f1 - f0)/(3*duration**2)
c = f0

# Sabendo que ax^2 + bx + c
f_quadratic = a*(time)**2 + c

quadratic_chirp = np.cos(2*pi*f_quadratic*time)

quad_chirp = chirp(time, f0=f0, f1=f1, t1=duration, method='quadratic')
```

```python
figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.plot(time, quadratic_chirp)
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Amplitude")
ax1.grid(True)

ax2.plot(time, quad_chirp)
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("Amplitude")
ax2.grid(True)

plt.show()
```

```python
logarithmic_chirp = chirp(time, f0=f0, f1=f1, t1=duration, method='logarithmic')
```

```python
plt.plot(time, logarithmic_chirp)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

### (b) Escreva um código para tocar os sinais gerados.

```python
# Audio do linear chirp

Audio(lin_chirp, rate=fs)
```

```python
# Audio do quadratic chirp

Audio(quad_chirp, rate=fs)
```

```python
# Audio do logarithmic chirp

Audio(logarithmic_chirp, rate=fs)
```
