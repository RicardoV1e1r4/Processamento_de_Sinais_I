```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from IPython.display import Audio
```

### (a) Gere um gráfico do sinal de áudio no domínio do tempo.

```python
samplerate, data = wavfile.read("/content/data_handel.wav")

# Número de amostras N
N = data.shape[0]

# Vetor de tempo
time = np.arange(N) / samplerate
# print(time)
# print(N)
# print(samplerate)
# print(data.shape)
```

```python
plt.figure(figsize=(12,6))
plt.plot(time, data)
plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
plt.grid(True)
plt.show()
```

### (b) Escreva um código para tocar os sinais gerados. Ouça o sinal de áudio com a frequência de amostragem original do sinal original, com o dobro da frequência original e o quádruplo da frequência original.

```python
# Áudio com a frequência original
Audio(data=data, rate=samplerate)
```

```python
# Áudio com o dobro da frequência original
Audio(data=data, rate=samplerate*2)
```

```python
# Áudio com quatro vezes frequência original
Audio(data=data, rate=samplerate*4)
```
