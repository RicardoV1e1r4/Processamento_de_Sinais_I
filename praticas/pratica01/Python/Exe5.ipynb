```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from IPython.display import Audio
```

```python
fs_h, h = wavfile.read('/content/data_h_banheiro.wav')
fs_t, taca = wavfile.read('data_sinal_taca.wav')
```

### (a) Gere um gráfico dos sinais no domínio do tempo.

```python
plt.figure(figsize=(12,4))
plt.plot(h)
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.title('Resposta ao impulso — banheiro')
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(taca)
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.title('Sinal da taça')
plt.show()
```

### (b) Escreva um código para tocar os sinais gerados.

```python
# Áudio do banheiro

Audio(h, rate=fs_h)
```

```python
# Áudio da taça

Audio(taca, rate=fs_t)
```
